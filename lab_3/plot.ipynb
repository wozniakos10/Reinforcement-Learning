```python
import matplotlib.pyplot as plt
from matplotlib.ticker import PercentFormatter
```

```python


def plot_function(x, y, title, xlabel, ylabel, filename):
    plt.figure(figsize=[8,8])
    plt.plot(x, y, "-o")
    plt.title(title)
    plt.grid(True)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)

    plt.ylim(bottom=0)  # Skala Y od zera
    plt.gca().yaxis.set_major_formatter(PercentFormatter())  # Dodaj znak %

    plt.savefig(f"plots/{filename}", dpi=300, bbox_inches='tight')
    plt.show()
```

```python
x = [0.05,0.1,0.2,0.3,0.5]
y = [22.8, 34.8, 44.4, 57.2, 64.8]
plot_function(x,y,"Procentowa ilość wygranych gracza A", "t_limit", "Procent wygranych gracza A", "ucb_time_comparison")
```

```python
x = [0.1,0.3, 1,2]
y = [51.6, 51.6, 44.8, 39.2]
plot_function(x,y,"Procentowa ilość wygranych gracza A", "c", "Procent wygranych gracza A", "ucb_c_coeff_comparison")
```

```python
x = [0.01, 0.05, 0.1, 0.2, 0.5]
y = [47.2, 52.2, 46.6, 50.8, 48.2]
plot_function(x,y,"Procentowa ilość wygranych gracza A", "epsilon", "Procent wygranych gracza A", "eps_greedy_comparison")
```
