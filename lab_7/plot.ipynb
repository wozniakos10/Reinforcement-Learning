```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import json
import pandas as pd
```

```python
with open("evaluate_data/lunar_lander/sac2e5.json") as f:
    sac_data2e5 = json.load(f)

with open("evaluate_data/lunar_lander/dt_sac_2e5.json") as f:
    dt_data2e5 = json.load(f)

with open("evaluate_data/lunar_lander/sac5e5.json") as f:
    sac_data5e5 = json.load(f)

with open("evaluate_data/lunar_lander/dt_sac_5e5.json") as f:
    dt_data5e5 = json.load(f)

with open("evaluate_data/lunar_lander/ppo2e6.json") as f:
    ppo2e6 = json.load(f)

with open("evaluate_data/lunar_lander/dt_ppo_2e6.json") as f:
    dtppo2e6 = json.load(f)
```

```python
df_sac2e5 = pd.DataFrame({
    "reward": sac_data2e5["rewards"],
    "model": "SAC 2e5"
})

df_dt2e5 = pd.DataFrame({
    "reward": dt_data2e5["rewards"],
    "model": "DT SAC 2e5"
})

df_sac5e5 = pd.DataFrame({
    "reward": sac_data5e5["rewards"],
    "model": "SAC 5e5"
})

df_dt5e5 = pd.DataFrame({
    "reward": dt_data2e5["rewards"],
    "model": "DT SAC 5e5"
})

df_ppo2e6 = pd.DataFrame({
    "reward": ppo2e6["rewards"],
    "model": "PPO 2e6"
})

df_dtppo2e6 = pd.DataFrame({
    "reward": dtppo2e6["rewards"],
    "model": "DT PPO 2e6"
})

df2e5 = pd.concat([df_sac2e5, df_dt2e5], ignore_index=True)
df5e5 = pd.concat([df_sac5e5, df_dt5e5], ignore_index=True)
df_ppo = pd.concat([df_ppo2e6, df_dtppo2e6], ignore_index=True)
```

```python
sns.histplot(data=df2e5, x="reward", hue="model", bins=30, element="step", stat="count", common_norm=False, alpha=0.3)

plt.title("Comparison of rewards")
plt.xlabel("Reward")
plt.ylabel("Count")
plt.savefig("evaluate_data/lunar_lander/histogram2e5.png")
plt.show()
```

```python
sns.histplot(data=df5e5, x="reward", hue="model", bins=30, element="step", stat="count", common_norm=False, alpha=0.3)

plt.title("Comparison of rewards")
plt.xlabel("Reward")
plt.ylabel("Count")
plt.savefig("evaluate_data/lunar_lander/histogram5e5.png")
plt.show()
```

```python
sns.histplot(data=df_ppo, x="reward", hue="model", bins=30, element="step", stat="count", common_norm=False, alpha=0.3)

plt.title("Comparison of rewards")
plt.xlabel("Reward")
plt.ylabel("Count")
plt.savefig("evaluate_data/lunar_lander/histogramppo2e6.png")
plt.show()
```
